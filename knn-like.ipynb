```python
import pandas as pd
pd.options.plotting.backend = "plotly"
df = pd.read_csv('data/PRPA_load_cleaned_mjw.csv',
                 comment='#',
                 index_col=0,
                 parse_dates=True)['2021-6-7':'2022-8-10']
df.loc[:,'i'] = range(len(df))
df
```

```python
k = 30
```

```python
dt = pd.Timestamp('2022-8-6 15:00')
i = df.loc[dt,'i']
i
```

```python
n_days = int(i/24)
n_days
```

```python
begin = i%24
begin
```

```python
df
```

```python
df = df.drop(columns=['i'])
```

```python
library = df.iloc[begin:i].values.reshape(n_days,24).T
library = pd.DataFrame(library,index=[x%24 for x in range(begin,begin+24)])
library.insert(0,'x',pd.NA)
library
```

```python
x = df.iloc[i:i+24]
peak_h = df.Load.loc[pd.date_range(dt.floor('D'),periods=24,freq='1H')].idxmax().hour
peak_h
```

```python
x
```

```python
library.x = x.values
library
```

```python
library.corr()['x'].drop('x').sort_values(ascending=False).head(5)
```

```python
days = library.corr()['x'].drop('x').sort_values(ascending=False).head(k).index.to_list()
days
```

```python
next_days = [d+1 for d in days]
```

```python
if n_days in next_days:
    next_days.remove(n_days)
next_days
```

```python
library[next_days].idxmax()
```

```python
peak_h_pred = library[next_days].idxmax().mean()
peak_h_pred
```

```python
library[next_days].idxmax().mode()
```
